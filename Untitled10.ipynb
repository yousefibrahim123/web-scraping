```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import json
url=requests.get("https://books.toscrape.com/").content

# src=url.content
soup=BeautifulSoup(url,"html.parser")

# img=soup.find_all("li",{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
book_list=[]
for book in soup.find_all("li",{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"}):
    img1=book.find("div",{"class","image_container"})
    title=book.find("h3").text.strip()
    price=book.find("p",{"class":"price_color"}).text.strip()
    rating=book.find_all("p")
    book_list.append([title,price,rating,img1])
book_list
```

```python
df =pd.DataFrame(book_list,columns=["Title","Price","Rating","Image"])
df
```

```python
df.to_csv("book.csv")
df.to_json("book.json")
```
